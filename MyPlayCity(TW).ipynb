# Задание 2(а)

```python
RPI['trade_date']= pd.to_datetime(RPI['Install Day']).dt.date
```

```python
RPI
```

```python
RPI['day_week']= pd.to_datetime(RPI['trade_date']).dt.dayofweek
```

Выделяем из даты дни недели, где 0 - понедельник, 6 - воскресенье.

```python
q = RPI.iloc[0:, [3, 5, 9, 15]]
```

```python
q
```

```python
q[(q['day_week'] == 5) | (q['day_week'] == 6)].mean()
```

```python
q[(q['day_week'] != 5) & (q['day_week'] != 6)].mean()
```

На сдедующий день после установки игры (Day 1) пользователи, которые установили игру в выходные принесли в среднем больше дохода, нежели пользователи которые установили игру в будни. в 4 и 8 дне это не зависело на итоговые доходы.

# Задание 2(б)

```python
z = RPI.iloc[0:, [3, 5, 9]]
```

```python
z
```

```python
Ad_Views_per_Day = Ad_Views_per_Day.rename(columns = {'Day 1' : 'Ad_Views_per_Day_1'})
```

```python
Ad_Views_per_Day
```

```python
v = pd.concat([z, Ad_Views_per_Day['Ad_Views_per_Day_1']], axis=1)
```

```python
v
```

```python
v.corr()
```

Из таблицы выше видно, что у просмотра рекламы с прибылью положительная кореляция сильно отличная от нуля что говорит нам о том, что чем выше просмотр рекламы тем больше прибыль.

# Задание 2(в)

```python
Retention_day_1 = Retention['Day 1'].apply(lambda x: x.replace('%', '')).astype('float') / 100
```

```python
Retention_day_1.name = 'Retention_day_1'
```

```python
Retention_day_1
```

```python
c = pd.concat([z, Retention_day_1], axis=1)
```

```python
c
```

```python
c.corr()
```

Из таблицы выше видно, что у доли вернувшихся в игру пользователей с прибылью положительная кореляция сильно отличная от нуля что говорит нам о том, что чем выше доля вернувшихся в игру пользователей тем больше прибыль. В этом случае кореляция лучше с точки зрения прибыли, чем в предъидущем случае.
